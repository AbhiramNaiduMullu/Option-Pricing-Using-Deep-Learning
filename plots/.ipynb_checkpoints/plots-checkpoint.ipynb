```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = np.loadtxt('train-losses.txt')
val = np.loadtxt('val-losses.txt')
```

```python
plt.plot(range(1, 121), np.log(train))
plt.plot(range(1, 121), np.log(val))
plt.xlabel('Epoch')
plt.ylabel('log(MSE)')
plt.legend(['Training Loss', 'Validation Loss'])
plt.title('MLP1 Loss')
plt.savefig('MLP1_Loss.png')
```
